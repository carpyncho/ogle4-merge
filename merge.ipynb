```python
ogle4 = pd.concat(col.values(), ignore_index=True)
ogle4["cls"] = ogle4.cls.apply(lambda c: (("RRLyr-" + c) if c.startswith("RR") else c))
ogle4["cat_cls"] = ogle4["cls"].copy()
ogle4 = ogle4[["id", "ra", "dec", "cat_cls"]].copy()
```

```python
OGLE4_TO_OGLE3 = {
    'CEP-1O': 'Cep-1',
    'CEP-1O/2O': 'Cep-12',
    'CEP-1O/2O/3O': 'Cep-123',
    'CEP-1O/3O': 'Cep-13',
    'CEP-2O': 'Cep-2',
    'CEP-2O/3O': 'Cep-23',
    'CEP-F': 'Cep-F',
    'CEP-F/1O': 'Cep-F1',
    'CEP-F/1O/2O': 'Cep-F12',
    'aCEP-1O': 'aCep-1',
    'aCEP-F': 'aCep-F',
    'aRRd': "aRRLyr-RRd"}

def change_cls(c):
    v = OGLE4_TO_OGLE3.get(c, c)
    return v

ogle4["cls"] = ogle4.cat_cls.apply(change_cls)
```
